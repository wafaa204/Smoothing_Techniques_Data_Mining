```python

import numpy as np
array=[3, 7, 8, 5, 12, 14, 21, 13, 18, 23, 27, 30]

#########if you need to make the user enter list of number and make it smooth ..
'''
array=[]
print("How many number you want to enter: ")
answer=int(input())
flag=True
while(flag):
    for i in range(answer):
        print(f"please enter number {i} : ")
        enter=int(input())
        array.append(enter)
    flag=False
'''
###################################################################
print("The original array : ",array)
#sort the array
array.sort()
print("Sorted array: ",array)
#create array to divide the data in it
bins=[]
#store the size of bin in variable named by size ..
#so if you want to change the size of group modify it this variable
size=3
#crate for loop to divide array in to group and each group contain size
for i in range(0, len(array), size):
   bins.append(array[i:i + size])
#create another for loop to print the groups as bins
for i in range(len(bins)):
        print(f"Bin {i+1} =",bins[i])
#create fuction to divide array .. later on you understand why we create this fuction
def create_bins(array,size):
    return [array[i:i+size] for i in range(0,len(array),size)]
#create function to find the mean for the data set by using method 'mean()'
def by_mean(bins):
    return [np.mean(i)for i in bins]
#create function to find the median for the data set by using method 'median()'
def by_median(bins):
    return[np.median(i)for i in bins]
#create function to find the boundaries for the data set
def by_boundaries(bins):
    #here if the minimum value equal the first index or if the
    #maximum value equal the last index return the first index else return last index
    return [bin[0] if (bin[0] == min(bin) or bin[-1] == max(bin))
    else bin[-1] for bin in bins]
bins=create_bins(array,size)
#create new array to store mean in it
mean_array=[]
#create new array to store median in it
median_array=[]
#create new array to store boundaries in it
boundaries_array=[]
#create for loop and use method 'extend()' to store the value in the array that had been created
for bin in bins:
    #we use method 'ceil()' to near the number for the up
    mean_array.extend([np.ceil(np.mean(bin))]*len(bin))
    median_array.extend([np.ceil(np.median(bin))]*len(bin))
    #create variable named bu min_value to store the minimum value in it by using method 'min()'
    min_value=min(bin)
    #create variable named bu max_value to store the maximum value in it by using method 'max()'
    max_value=max(bin)
    #substract minimum value from bin(elements of bins) and substract maximum value from
    #bin(elements of bins)to know the difference between number if the  deference of minimum value
    #less than maximum value return min_value else return maximum value
    boundaries_array.extend([min_value if np.ceil(abs(min_value - i))< np.ceil(abs(max_value - i)) else max_value for i in bin])
#create count because when i print the statement show number of bin
count=1
print("\nSMOOTHING BY MEAN :")
#use for loop print element from i to the size that have been defined above
for i in range (0,len(array),3):
    print(f"Bin {count} =",mean_array[i:i+size])
    count+=1 #increase one step for every loop
#use for loop print element from i to the size that have been defined above
count=1
print("\nSMOOTHING BY MEDIAN :")
for i in range (0,len(array),3):
    print(f"Bin {count} =",median_array[i:i+size])
    count+=1
#use for loop print element from i to the size that have been defined above
count=1
print("\nSMOOTHING BY BOUNDRIES :")
for i in range (0,len(array),3):
    print(f"Bin {count} =",boundaries_array[i:i+size])
    count+=1

```
